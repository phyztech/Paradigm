```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
pathents_in_russia = pd.read_csv("pathents_in_russia.csv", header=1)
```

```python
pathents_in_russia['priority date'] = pathents_in_russia['priority date'].apply(lambda x: datetime.date(*map(int, x.split("-"))))
```

```python
year_hist = pathents_in_russia["priority date"].apply(lambda x: x.year).values
```

```python
plt.figure(figsize=(16, 9))
plt.hist(year_hist, bins=len(np.unique(year_hist)) - 1, edgecolor="k")
plt.title("Distribution of Russian patents connected with blockchain", fontsize=20)
plt.ylabel("Count", fontsize=16)
plt.xlabel("Year", fontsize=16)
plt.show()
```

```python
records = pd.read_csv("./records.csv")
```

```python
records.iloc[0]
```

```python
data = records["TI"].values
```

```python
from nltk.tokenize import TweetTokenizer
tokenizer = TweetTokenizer()
preprocess = lambda text: tokenizer.tokenize(text.lower())
```

```python
data_prep = list(map(preprocess, data))
```

```python
data_ = []
for row in data_prep:
    data_.extend(row)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.colors as mcolors
```

```python
stop = STOPWORDS
```

```python
cols = [color for name, color in mcolors.TABLEAU_COLORS.items()]  # more colors: 'mcolors.XKCD_COLORS'
```

```python
un_data, counts = np.unique(data_, return_counts=True)
dict_data = dict(zip(un_data, counts))
```

```python
stop = set.union(stop, {
    ",", "?", ":", "using", "(", ")", "-", "systems", "based", 
    "research", "analysis", "networks", "new", "supply", "approach", "model"})

for stop_word in stop:
    if stop_word in dict_data:
        dict_data.pop(stop_word)
```

```python
cloud = WordCloud(stopwords=stop,
                  background_color='white',
                  width=2500,
                  height=1800,
                  max_words=30,
                  colormap='tab20')
```

```python
cloud.generate_from_frequencies(dict_data, max_font_size=300)
```

```python
plt.figure(figsize=(16, 9))
plt.imshow(cloud)
```

```python
plt.imsave("./map.jpg", cloud.to_array())
```
